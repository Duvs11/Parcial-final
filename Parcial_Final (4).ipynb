```python
gdf_barrios=gpd.read_file("/content/drive/MyDrive/POO/Barrios_S.geojson")
gdf_comuna_3 = gdf_barrios.loc[gdf_barrios['COM_MAPA'] == '7'].copy()
gdf_comuna_3.to_crs(epsg=9377,inplace=True)
gdf_comuna_3["área(m^2)"]=gdf_comuna_3.area
gdf_comuna_3.head()
```

```python
gdf_comuna_3['área(m^2)STR'] = gdf_comuna_3['área(m^2)'].astype(str)
gdf_comuna_3
```

```python
mapaC=gdf_comuna_3.plot(
    column="área(m^2)STR",
    figsize=(15,7.5),
    edgecolor='black',
    cmap='gist_rainbow',
    legend=True,
legend_kwds={
    'loc':'upper left',
    'bbox_to_anchor':(1,1),
    'title':'Áreas',
    'ncol':6
}
)
ctx.add_basemap(mapaC, crs=gdf_comuna_3.crs.to_string())
```
